```python
import time
import keyboard
import pyautogui
from tft_auto import getImages, enterGame, ff, restart, is_three_one

# Will contiunously keep the loop going
cont = True
# Extra time for the user to get the League of Legends client to the game lobby
time.sleep(5)
# Start the queue
pyautogui.click(900, 850)
while cont:
    
    # Will check if we are on stage 2-1
    three_one = False
    
    while not three_one:    
        # Will repeatedly try to get into the game by accepting the queue
        enterGame()
    
        # Checks if we get to stage 3-1 and will exit this loop
        three_one = is_three_one()
        
    # Forfeits the game
    ff()
    
    # Enter the queue for the name game
    restart()
```
